```python
import matplotlib.pyplot as plt
import numpy as np

ab_list = open("antibiotic_list.txt","r")
ab_list = ab_list.read().splitlines()

############################################
selected_group = 'Penicillins'
mode = 'hardft'


if mode == 'all':
    data = dataALL
elif mode == 'one':
    data = dataONE
elif mode == 'three':
    data = dataTHREE
elif mode == 'easyft':
    data = dataEASYFT
elif mode == 'hardft':
    data = dataHARDFT
sensitivity = pd.DataFrame(data)

df = sensitivity
type = 'Sensitivity'
if mode == 'easyft':
   mask = '30%'
elif mode == 'hardft':
   mask = '60%'

selected_rows = group_indices[selected_group]

selected_data = df.iloc[selected_rows]

df_transposed = selected_data.T

antibiotic_names = [ab_list[i] for i in selected_rows]

plt.figure(figsize=(12, 6))

positions = np.arange(len(df_transposed.columns))

bar_width = 0.15  # Adjust the width here

num_groups = len(df_transposed)

# Define gradients for blues, greens, and oranges within each group
gradient_length = 3  # Number of gradient steps
blues = plt.cm.Greens(np.linspace(0.5, 1, gradient_length))  # Linear gradient from light to dark
greens = plt.cm.Blues(np.linspace(0.5, 1, gradient_length))  # Linear gradient from light to dark
oranges = plt.cm.Purples(np.linspace(0.5, 1, gradient_length))  # Linear gradient from light to dark

# Flatten the gradients to get a single list of colors
group_colors = [color for gradient in zip(blues, greens, oranges) for color in gradient] * 2  # Repeat for each group

# Iterate through each row (antibiotic) and plot its bars with gradient colors within each group
for i, (antibiotic, row) in enumerate(df_transposed.iterrows()):
    # Assign gradient colors to bars within each group
    colors_within_group = group_colors[i % (3 * gradient_length)::(3 * gradient_length)]  # Select colors for each bar within the group
    plt.bar(positions + i * bar_width, row, width=bar_width, label=antibiotic, color=colors_within_group)

plt.xlabel('Antibiotic')
plt.ylabel(type)
plt.title(type + ' of ' + selected_group + ' with different encoders ('+mask+' Masking)')
plt.xticks(positions + (num_groups - 1) * bar_width / 2, antibiotic_names, rotation=45, ha='right')
plt.legend(title='Antibiotic')
plt.tight_layout()
plt.show()
```
